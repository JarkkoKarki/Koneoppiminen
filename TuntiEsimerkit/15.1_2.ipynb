```python
def plot_data(df):
    plt.figure(figsize=(10, 10))
    for i, col in enumerate(df.select_dtypes(include=['int', 'float']).columns):
        plt.subplot(3,3, i + 1)
        df[col].plot(kind='hist', title=col)
        plt.xlabel(col)
plot_data(df)
```
