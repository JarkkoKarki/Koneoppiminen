$$ 
y = \beta_{0} + \beta_{1} x
$$

$$ 
ennuste = \beta_{0} + \beta_{1} x^i  , Oikea= y^i
$$

Pienimmän neliösumman menetelmä (MSE)

$$ 
min \beta_{0}\beta_{1}(\sum_{i=1} (\beta_{0} + \beta_{1} x^i - y^i)^2)

\\n
(\sum_{i=1}= opetusjoukko)
$$



Suoritus kyvyn mittaus
Mean Absolute Error (MAE)
$$ 
\sum_{i=1} \vert \beta_{0} + \beta_{1} x^i-y^i \vert
$$


(MSE) usealla muuttujalla

$$ 
min \beta_{0}\beta_{1}(\sum_{i=1} (\beta_{0} + \beta_{1} x_{1}^i +  \beta_{2} x_{2}^i + \beta_{n} x_{n}^i - y^i)^2)
$$
